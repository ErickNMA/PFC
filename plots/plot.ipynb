```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
from scipy.signal import find_peaks

# Configurações de plot:
plt.style.use([
    'grid',
    'retro'
])
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
#txt = np.loadtxt('data.txt')   #t	j1	j2	j3	j4	j5	grip	ref1	ref2	ref3	ref4	ref5	refg	u1	u2	u3	u4	u5	u6
#txt = np.loadtxt('j1.txt')
#txt = np.loadtxt('j2.txt')
#txt = np.loadtxt('j3.txt')
#txt = np.loadtxt('grip.txt')
txt = np.loadtxt('movejntrapido.txt')
data = []
for i in range(len(txt[0])):
    data.append((txt[:,i])[50:350])
    #data.append(txt[:,i])
#data[0] -= 13
```

```python
ticks = [[[0, -90], [0, -90]], [[0, -30], [0, -30]], [[0, -60], [0, -60]], [[0, 90], [0, 90]], [[-5, 5], [-5, 5]], [[0, 80], [0, 80]]]

for i in range(6):
    plt.subplots(2,1, sharex=True)
    plt.subplots_adjust(hspace=0.05)
    plt.subplot(2,1,1)
    #plt.title(f'GDL {i+1}')
    plt.plot(data[0], data[i+7], linestyle='dotted', color='C0', label='$\\theta_{ref}(t)$')
    plt.plot(data[0], data[i+1], color='C2', label=f'$\\theta_{i+1}(t)$')
    plt.ylabel('Ângulo [°]')
    #plt.yticks([data[i+7][0], data[i+7][-1]], [data[i+7][0], data[i+7][-1]])
    plt.yticks(ticks[i][0], ticks[i][1])
    plt.legend()
    plt.subplot(2,1,2)
    plt.plot(data[0], data[i+13], color='C4')
    plt.ylabel('PWM [%]')
    plt.xlabel('Tempo [s]')
    plt.savefig(f'traj{i+1}.eps', dpi=600, transparent=True, bbox_inches='tight')
```
